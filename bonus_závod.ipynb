```python
# převedení na jednotný čas (minuty)
clean_set = []

for person in data:
    separate_times = person[1].split(';')
    final = []

    final.append(person[0])
    clean_set.append(final)
    for one in separate_times:
        minutes = one[:2].replace(':','')
        seconds = one[2:].replace(':','')
        seconds = round((float(seconds) / 60), 2)
        time = float(minutes) + seconds
        final.append(time)
print(clean_set)
```

```python
# rozdíl mezi prvním a posledním kolem
for one in clean_set:
    difference = round((one[-1] - one[1]), 2)
    if one[-1] > one[1]:
        plus = '+'
    else:
        plus = ''
    print(f' Rozdíl mezi časy prvního běžce a běžce č. {one[0]} je: {plus}{difference} minut.')
```

```python
# nejpomalejší a nejrychlejší kolo z prvních 5 kol závodu

for one in clean_set:
    first_five = one[1:6]
    print(f'V prvních 5 kolech u běžce č. {one[0]} byl rozdíl mezi nejrychlejším a nejpomejším kolem {round((max(first_five) - min(first_five)), 2)} minut.') 
```

```python
# nejpomalejší a nejrychlejší kolo z celého závodu

for one in clean_set:
    wanted = one[1:]
    print(f'Běžec č. {one[0]} měl nejrychlejší {wanted.index(max(wanted)) + 1}. kolo a nejpomalejší {wanted.index(min(wanted)) + 1}. kolo.')
```

```python
# průměrná rychlost běhu 

for one in clean_set:
    wanted = one[1:]
    total_time_hours = sum(wanted) / 60
    average_time = 42.195 / total_time_hours
    print(f'Běžec č. {one[0]} běžel během závodu průměrně rychlostí {round(average_time, 3)} km/h .')
```

```python
# kolik běžců a běžkyň zaběhlo nejrychleji první a nejpomaleji poslední kolo běhu?
fastest = 0
slowest = 0
for one in clean_set:
    wanted = one[1:]
    if (min(wanted)) == wanted[0]:
        fastest += 1
    if (max(wanted)) == wanted[-1]:
        slowest  += 1
print(f"Nejrychlejší první kolo mělo {fastest} účastníků běhu. Činí to {round((fastest/len(clean_set)),3) * 100} % z účastníků.")
print(f"Nejrychlejší první kolo mělo {slowest} účastníků běhu. Činí to {round((slowest/len(clean_set)),3) * 100} % z účastníků.")
```
